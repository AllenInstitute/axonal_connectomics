```python
# jupyter notebook implementing fixcurvedstack with pixel-by-pixel shifts determined from ImageJ macro filters
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

from skimage import io
from skimage import util as skutil
import os
#%matplotlib inline
#%matplotlib notebook
%load_ext autoreload
%autoreload 2
```

```python
import kt_utils_v4 as kt # kt_utils_v2.py must be in the same directory
import cr_utils_v2 as cr # cr_utils_v2.py must be in the same directory
datestring = '0427' # datestring is appended to saved files to indicate version of generating code
```

```python
# read a tiff stack
#datadir = '.' # replace this string with tif directory
datadir = '.'
fileroot = 'tissuecyte_sections20x_s3_box1_G_Short_zm4'
infile = os.path.join(datadir,fileroot+'.tif')

print('infile_name ' + infile)
dsstack = io.imread(infile)
print('read imstack ' + infile)
```

```python
max_radius = 16 # for 4x downsampled to 512x512
smooth_radius = 32 # for 4x downsampled
```

```python
fig=plt.figure(figsize=(10,10))
plt.imshow(skutil.montage(dsstack,fill='mean'),cmap='pink')
plt.axis('off')
cbar = plt.colorbar(shrink=.8)
cbar.ax.locator_params(nbins=40)
```

```python
# NOTE: this takes awhile
fstack = kt.applyIJFiltersStack(dsstack,max_radius,smooth_radius)
```

```python
fmont = skutil.montage(fstack,fill=0)
fig=plt.figure(figsize=(10,10))
plt.imshow(fmont,cmap='rainbow')
plt.axis('off')
cbar = plt.colorbar(shrink=.8)
cbar.ax.locator_params(nbins=40)
```

```python
minMaxImage = kt.minMaxThreshold(fstack)
plt.imshow(minMaxImage,cmap='rainbow')
plt.axis('off')
cbar = plt.colorbar(shrink=.8)
cbar.ax.locator_params(nbins=10)
```

```python
minImage = kt.medMinThreshold(fstack)
plt.imshow(minImage,cmap='rainbow')
plt.axis('off')
cbar = plt.colorbar(shrink=.8)
cbar.ax.locator_params(nbins=10)
```

```python
threshold = 200 # NOTE: changed from 40 to get expected output mask - is this a uint8 vs uint16 issue?
maskstack = kt.makeIJMask(fstack,threshold)
plt.subplot(1,2,1)
plt.imshow(np.max(maskstack,axis=0),'gray',vmin=0,vmax=1)
plt.subplot(1,2,2)
plt.imshow(np.min(maskstack,axis=0),'gray',vmin=0,vmax=1)
```

```python
fig=plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(skutil.montage(1*maskstack,fill=0),cmap='gray')
plt.subplot(1,2,2)
plt.imshow(fmont,cmap='rainbow')
```

```python
# optional save maskstack as tiff for sharing
outstring='_masks_' + datestring
maskfile=os.path.join(datadir,fileroot+outstring+'.tif')

kt.saveMaskStack(maskfile,maskstack)
print('saved mask stack to ' + maskfile)
```

```python
Zstart,Zend = kt.getFirstLastIndices(maskstack)
```

```python
# show beginning of data, end of data (top and bottom), and thickness
fig=plt.figure(figsize=(20,12))
plt.subplot(1,3,1)
plt.imshow(Zstart,cmap='rainbow') ; plt.axis('off') ; plt.title('Zstart') ; cbar = plt.colorbar(shrink=.4)
plt.subplot(1,3,2)
plt.imshow(Zend,cmap='rainbow') ; plt.axis('off') ; plt.title('Zend') ; cbar = plt.colorbar(shrink=.4)
plt.subplot(1,3,3)
plt.imshow(Zend-Zstart,cmap='rainbow') ; plt.axis('off') ; plt.title('Thickness') ; cbar = plt.colorbar(shrink=.4)
```

```python
dsstackflat = kt.fixcurvedstack(dsstack,Zstart,Zend)
```

```python
fig=plt.figure(figsize=(10,10))
plt.imshow(skutil.montage(dsstackflat,fill=100),cmap='pink')
plt.axis('off')
cbar = plt.colorbar(shrink=.8)
cbar.ax.locator_params(nbins=40)
```

```python
dsstackflat = kt.fixcurvedstack_interp(dsstack,Zstart,Zend)
```

```python
fig=plt.figure(figsize=(10,10))
plt.imshow(skutil.montage(dsstackflat,fill=100),cmap='pink')
plt.axis('off')
cbar = plt.colorbar(shrink=.8)
cbar.ax.locator_params(nbins=40)
```

```python
print(dsstackflat.dtype)
plt.subplot(1,2,1)
plt.imshow(dsstackflat[0,:,:],'gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(dsstack[0,:,:],'gray')
plt.axis('off')
```

```python
plt.subplot(1,2,1)
plt.imshow(np.max(dsstackflat,axis=0),'gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(np.max(dsstack,axis=0),'gray')
plt.axis('off')
```

```python
# save flattened downsampled stack for convenience
outstring='_flat_' + datestring
outfile=os.path.join(datadir,fileroot+outstring+'.tif')

io.imsave(outfile,dsstackflat)
print('saved flattened stack ' + outfile)
```
